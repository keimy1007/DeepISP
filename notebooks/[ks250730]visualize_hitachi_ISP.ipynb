# 日立ISPデータの可視化

整理した日立ISPデータ（isp_binocular_data.csv）から10症例を選んで視野をプロットする。

```python
import pandas as pd
import numpy as np
# フォントを一括設定
import matplotlib.pyplot as plt
# import matplotlib.font_manager as fm
import matplotlib
matplotlib.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['font.weight'] = 'bold'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.titlesize'] = 12
# matplotlib.rcParams['font.style'] = 'italic'


import os
import sys
sys.path.append('..')

from src.visualize import visualize_isp_sample

# 日本語フォントの設定
plt.rcParams['font.family'] = 'DejaVu Sans'

# ルートに移動
%cd /Users/keimy/git/DeepISP
%pwd
```

```python
# データの読み込み
print("=== データの読み込み ===")

# 日立ISPデータ（前処理済み）
df_isp = pd.read_csv('outputs/hitachi_isp/isp_binocular_data.csv')
print(f"ISPデータ: {len(df_isp)}行")

# ISP座標データ
coord_isp = pd.read_csv('datasets/solid/coord_isp.csv')
print(f"ISP座標: {len(coord_isp)}点")

# データの確認
print("\nサンプルデータ:")
print(df_isp.head())
```

```python
n_sample = 100
unique_patients = df_isp['daily_id'].unique()
np.random.seed(42)  # 再現性のため
selected_patients = np.random.choice(unique_patients, size=min(n_sample, len(unique_patients)), replace=False)

print(f"選択された患者ID: {selected_patients}")
```

```python
# visualize_isp_sampleを使用して10症例を個別にプロット
os.makedirs('figures/hitachi_isp_plots', exist_ok=True)

for i, patient_id in enumerate(selected_patients[:n_sample]):
    patient_data = df_isp[df_isp['daily_id'] == patient_id]
    
    # 右眼のデータ
    od_data = patient_data[patient_data['eye'] == 'OD']
    if len(od_data) > 0:
        # point1〜point28の値を取得
        isp_plot = [od_data.iloc[0][f'point{i+1}'] for i in range(28)]
        # NaNを0に変換
        isp_plot = [int(v) if pd.notna(v) else 0 for v in isp_plot]
        
        # visualize_isp_sampleを呼び出し
        visualize_isp_sample(isp_plot, 
                           figname=f"Patient_{patient_id}_OD", 
                           save_path="figures/hitachi_isp_plots", 
                           color1="blue", color0="red")
        plt.title(f"Patient {patient_id} - OD (右眼)", fontsize=15)
        plt.show()
        
    # 左眼のデータ  
    os_data = patient_data[patient_data['eye'] == 'OS']
    if len(os_data) > 0:
        # point1〜point28の値を取得
        isp_plot = [os_data.iloc[0][f'point{i+1}'] for i in range(28)]
        # NaNを0に変換
        isp_plot = [int(v) if pd.notna(v) else 0 for v in isp_plot]
        
        visualize_isp_sample(isp_plot, 
                           figname=f"Patient_{patient_id}_OS", 
                           save_path="figures/hitachi_isp_plots", 
                           color1="blue", color0="red")
        plt.title(f"Patient {patient_id} - OS (左眼)", fontsize=15)
        plt.show()

print(f"\\n10症例の視野をプロットしました。")
```

```python
# 統計情報の表示
print("=== 選択された症例の統計 ===")

for patient_id in selected_patients[:5]:  # 最初の5症例のみ詳細表示
    patient_data = df_isp[df_isp['daily_id'] == patient_id]
    print(f"\n患者ID: {patient_id}")
    
    for eye in ['OD', 'OS']:
        eye_data = patient_data[patient_data['eye'] == eye]
        if len(eye_data) > 0:
            data_row = eye_data.iloc[0]
            
            # 各測定点の値を集計
            point_cols = [f'point{i+1}' for i in range(28)]
            values = [data_row[col] for col in point_cols if col in data_row]
            valid_values = [v for v in values if pd.notna(v)]
            
            if valid_values:
                visible_count = sum(v == 1 for v in valid_values)
                invisible_count = sum(v == 0 for v in valid_values)
                eye_label = '右眼' if eye == 'OD' else '左眼'
                
                print(f"  {eye_label}: 見えた={visible_count}/28, 見えない={invisible_count}/28")
```
